```python
data_prod = pd.read_csv(
    r"/home/pdconte/Desktop/vm_app/data/produccin-de-pozos-de-gas-y-petrleo-no-convencional.csv"
)
```

```python
data_prod.head()
```

```python
data_prod.columns
```

```python
prod_columns=['idempresa', 'anio', 'mes', 'idpozo', 'prod_pet', 'prod_gas',
       'prod_agua', 'iny_agua', 'tipopozo',
       'fechaingreso', 'empresa', 'sigla', 'profundidad', 'formacion',
       'areayacimiento', 'coordenadax', 'coordenaday', 'tipo_de_recurso',
       'sub_tipo_recurso'
       ]
```

```python
data_prod = data_prod[prod_columns]
```

```python
data_prod.rename(columns={'idempresa':'company_id',
                          'anio':'year',
                          'mes':'month',
                          'idpozo':'well_id',
                          'prod_pet': 'oil_prod_m3',
                          'prod_gas':'gas_prod_km3',
                          'prod_agua':'water_prod_m3',
                          'iny_agua':'water_inj_m3',
                          'tipopozo': 'well_type',
                          'fechaingreso': 'entry_date',
                          'empresa': 'company',
                          'sigla': 'well_name',
                          'profundidad': 'depth_m',
                          'formacion': 'formation',
                          'areayacimiento': 'field',
                          'coordenadax': 'Xcoor',
                          'coordenaday': 'Ycoor',
                          'tipo_de_recurso': 'type',
                          'sub_tipo_recurso': 'subtype'
                          }, inplace=True)
```

```python
data_prod.columns
```

```python
filtered_data_prod = data_prod.loc[
    (data_prod["formation"] == "vaca muerta")
    & (data_prod["well_type"].isin(["Petrolífero", "Gasífero", "Otro tipo"]))
    & (data_prod["type"] == "NO CONVENCIONAL")
    & (data_prod["subtype"] == "SHALE")
].copy()
```

```python
filtered_data_prod["month_count"] = (
    filtered_data_prod.groupby(["well_id"])["month"].cumcount() + 1
)
filtered_data_prod["oil_cum_m3"] = filtered_data_prod.groupby("well_id")[
    "oil_prod_m3"
].cumsum()
filtered_data_prod["gas_cum_km3"] = filtered_data_prod.groupby("well_id")[
    "gas_prod_km3"
].cumsum()
filtered_data_prod["water_prod_cum_m3"] = filtered_data_prod.groupby("well_id")[
    "water_prod_m3"
].cumsum()
filtered_data_prod["water_inj_cum_m3"] = filtered_data_prod.groupby("well_id")[
    "water_inj_m3"
].cumsum()
```

```python
filtered_data_prod = filtered_data_prod.sort_values(["well_id", "year", "month"])
filtered_data_prod
```

```python
filtered_data_prod.drop(columns=['formation'], inplace=True)
```

```python
# Get the list of well IDs from filtered fracturing data
well_ids_in_frac = filtered_data_frac["well_id"].unique()

# Filter production data to only include wells that are in the fracturing data
filtered_data_prod = filtered_data_prod[data_prod["well_id"].isin(well_ids_in_frac)]
```
